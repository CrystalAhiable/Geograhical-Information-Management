**1.1.**<p style='text-align: justify;'> In detecting the extent of flood using remote sensing technologies, several data products can be considered. Water displays different characteristics at different wavelengths and knowledge in this is important when choosing a remote sensing technology to implement. For instance, in the shorter visible wavelengths (Visible), water has a better reflectance and water appears brighter than in the longer wavelength regions (e.g. NIR, Microwave) where more absorption of water occurs causing it to appear darker. <p style='text-align: justify;'> However, taking into consideration cloud build-up highly associated with floods, the Sentinel-1 SAR constellation will be used to examine the extent of flood in the coastal and southern areas of Myanmar. <p style='text-align: justify;'> When energy interacts with a surface object, it is either transmitted, absorbed, or reflected. These interactions vary depending on the wavelengths. For remote sensing purposes, reflectance is measured in the visible and near infrared region while in the microwave region, the backscatter coefficient is measured. The backscatter refers to the wave signals that are bounced back to the sensor after interacting with surface objects. Its intensity is represented in each pixel of the image and will vary depending on the characteristics of the surface feature, causing features to appear light or dark. On rougher surfaces, signal is scattered in all directions returning a significant amount to the sensor, hence, features appear to be brighter. On smoother surfaces, signals are scattered away from the sensor and features appear darker. In this analysis, flooded areas will be characterised by stagnant water, hence, it will cause a specular reflectance of waves, resulting in a low backscatter coefficient.</p>

<p style='text-align: justify;'> To obtain the backscatter coefficient from images, digital numbers in the imagery are converted using the sensor calibration coefficient A. The conversion is done using equation 1 where the raw image represents the DN and A is the provided in the metadata file of the product.</p>

$$        \sigma^o = \frac{DN^2}{A_{\sigma}^2}........................(Equation 1)$$

**1.3.**<p style='text-align: justify;'> As a result of the side-looking nature of Radar sensors, images develop some distortions in the form of relief displacement. Foreshortening occurs when waves oscillating from the sensor arrive at the base of a tall feature before hitting the top (A and B) resulting in the image appearing compressed and misrepresents the actual slope on ground ($A^{I}$ and $B^{I}$). Similar to foreshortening is the layover effect which occurs when the oscillating waves reach the top of the slope before the base (H and G), causing a layover of the top over the base resulting in a displacement from its original position towards the sensor ($H^{I}$ and $G^{I}$). Shadows occur when oscillating waves are unable to interact with and illuminate some parts of ground features due to the steepness of the slope, resulting in a shadowing effect which causes unilluminated areas to appear darker in radar images (Braun and Hochschild, 2017). This limitation is of particular relevance to flood mapping since those dark areas are likely to interfere with flood analysis.</p>



*Figure 3: Effects of Topography on SAR Images (Braun and Hochschild, 2017)*


<p style='text-align: justify;'> Speckle occurs due to differences in reflectance from adjacent pixels resulting in a salt and pepper effect causing a distortion in the image quality in the form of noise (Figure 4). </p>



*Figure 4: Speckle distortion on a SAR image showing the grainy effect caused by variance of negative and positive interference of pixel brightness (ESA, 2020)*

<p style='text-align: justify;'> These limitations can be reduced by a series of pre-processing steps. Distortions by effects of Topography can be reduced through geometric correction. This can be done using a digital elevation model and some ground control points to orthorectify the image (Chen & Wang, 2008). Shadow effects can also be reduced by obtaining different images from different angles.
In addition, speckle removal processes such as filtering can be applied to the image to reduce the noise and improve the image quality. However, it must be done carefully to prevent over-smoothening the signals in the image (Nezry, 2014) </p>

**3.1.**<p style='text-align: justify;'> The pixel-based supervised classification method was selected for producing the crop map. This method of classification involves the use of training datasets as a basis for the classification of the image. Based on the spectral pattern of the trained data, a computer algorithm is applied to determine the most possible class for each pixel of the image. The Maximum Likelihood Classification algorithm was used. This algorithm takes into consideration the variance and covariance of the spectral pattern when assigning a class to a pixel (Lillesand et al, 2015). The classification of each pixel is assigned based on its maximum probability of belonging to a particular class. </p>

<p style='text-align: justify;'> The supervised classification method was chosen because it offers the producer much flexibility to select informational classes to suit a particular purpose, hence, giving a higher tendency to achieve better accuracy results compared to using an unsupervised method (Campbell & Wynne, 2011). Considering the user had pre-defined the required classes, a supervised classification approach is deemed fit to accurately identify each of the information classes. Also, due to the images low-resolution, a pixel-based classification approach is preferred over the object-based classification approach which is suitable for high-resolution imagery (Ma, et al., 2017). </p>

<p style='text-align: justify;'> The Random forest (RF) classifier was chosen to replicate the analysis carried out by (Maxwell, Warner, & Fang, 2018). For this methodology, a split size of 50/50 was used instead of the 25/75 train-test-split size used in the article. This was done as an attempt to prevent an under-sampling of minority classes in order to improve the overall accuracy. An analysis by (Millard & Richardson 2015) suggests that the larger the training sample, the higher the likelihood of achieving better accuracy results. Many authors have debated on the ideal split size and suggest that it is common practise to use unbalanced split ratios (e.g. 80/20, 67/33). However, (Brownlee, 2020) suggests a split size of 50/50 would be ideal, hence, the basis for choosing a train-test-split ratio of 50/50.

```python
import gdal
import numpy as np
import pandas as pd
```

```python
#load image as array
original_image = gdal.Open('Assignment/aviris_hyperspectral_data/19920612_AVIRIS_IndianPine_Site3.tif')
image_pixels = original_image.ReadAsArray()
image_pixels.shape
```

```python
#reshape image pixels
nbands, nrows, ncols = image_pixels.shape
row_value = image_pixels.transpose(1, 2, 0).reshape((nrows*ncols, nbands))
np.testing.assert_array_equal(row_value[2], image_pixels[:, 0, 2])
print(row_value.shape)
```

```python
#load clr file
clr_data = pd.read_csv('Assignment/documentation/Site3_Project_and_Ground_Reference_Files/19920612_AVIRIS_IndianPine_Site3_gr.clr', header=None, delim_whitespace=True, names= ["A", "B", "C", "D", "E"]) 

#to clean-up label in clr table
for i in range(len(clr_data)):
    clr_data.iloc[i,4][2:][:-1]
```

```python
#load reference image as array
ref_data = gdal.Open('Assignment/documentation/Site3_Project_and_Ground_Reference_Files/19920612_AVIRIS_IndianPine_Site3_gr.tif')
ref_data_pixels = ref_data.ReadAsArray()
ref_data_pixels.shape
```

```python
nrows, ncols = ref_data_pixels.shape
row_val_ref_data = ref_data_pixels.reshape((nrows*ncols, 1))
np.testing.assert_array_equal(row_val_ref_data[1], ref_data_pixels[0, 1])
print(row_val_ref_data.shape)
```

```python
#recode values in ground reference data
ref_dict = {0:0, 1:1, 2:2, 3:2, 4:2, 5:3, 6:3, 7:3, 8:4, 9:5, 10:6, 11:6, 12:6, 13:7, 14:8, 15:8, 16:0}

row_value_ref = []
for i in row_val_ref_data:
    row_value_ref.append([ref_dict[i[0]]])
    
#convert from list to array
row_value_ref_arr = np.array(row_value_ref)
```

```python
new_table = np.zeros((21025, 221))
new_table.shape
```

```python
new_table = np.empty((21025, 221),dtype=float)
new_table[:,:-1] = row_value
new_table[:,-1:] = row_value_ref_arr
```

```python
#eliminate labels with zero values (not part of the reference data)
new_array = []

for i in range(len(new_table)):
    if new_table[i, -1] !=0:
        new_array.append(new_table[i, :])
```

```python
ground_reference = pd.DataFrame(new_array) 

ground_reference.head()
```

```python
#train-test-split
from sklearn.model_selection import train_test_split
x_train, x_validate, y_train, y_validate = train_test_split( 
    ground_reference.iloc[:, 0:220],
    ground_reference.iloc[:, 220],
    stratify=ground_reference.iloc[:, 220],
    test_size=0.50,
    random_state=42
)

print(x_train.shape)
print(x_validate.shape)
print(y_train.shape)
print(y_validate.shape)
```

```python
#variable selection using recursive feature elimination

from sklearn.datasets import make_classification
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier

rfe = RFE(estimator=DecisionTreeClassifier(), n_features_to_select=33)

rfe.fit(x_train, y_train)

for i in range(x_train.shape[1]):
    'Column: %d, Selected %s, Rank: %.3f' % (i, rfe.support_[i], rfe.ranking_[i])
    
x_training = x_train.loc[:, rfe.support_]
x_validation = x_validate.loc[:, rfe.support_]
print(x_training.shape)
print(x_validation.shape)
```

```python
#hyper-parameter selection

from sklearn import model_selection
from sklearn import ensemble
param_grid = {
    'bootstrap': [True],
    'max_depth': [8],
    'max_features': [33],
    'min_samples_leaf': [6, 7, 8],
    'min_samples_split': [7],
    'n_estimators': [200, 250]
}
rf = ensemble.RandomForestClassifier()
grid_search = model_selection.GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(x_training, y_train)
params = grid_search.best_params_
params
```

```python
rforest = ensemble.RandomForestClassifier(
    bootstrap=params['bootstrap'],
    max_depth=params['max_depth'],
    max_features=params['max_features'],
    min_samples_leaf=params['min_samples_leaf'],
    min_samples_split=params['min_samples_split'],
    n_estimators=params['n_estimators'],
    random_state=42
)
rforest.fit(x_training, y_train);
```

```python
rpred = rforest.predict(x_validation)
```

```python
from sklearn.metrics import confusion_matrix
all_classes = ground_reference[220].unique()

cm = confusion_matrix(
    y_validate,
    rpred,
    labels=all_classes
    
)

print(cm)
```

```python
#normalize confusion matrix

cmn = cm/21025
print(cmn)
```

```python
#change labels to string

new_dict = {0:'Other', 1:'Alfalfa', 2:'Corn', 3:'Grass', 4:'Hay', 5:'Oats', 6:'Soybeans', 7:'Wheat', 8:'Trees'}

#convert to list

all_classes_new = []
for i in all_classes:
    all_classes_new.append(new_dict[i])
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

fig, ax = plt.subplots()
ax.imshow(cmn, interpolation='nearest', cmap=plt.cm.Reds)

ax.set_xticks(np.arange(len(all_classes_new)))
ax.set_yticks(np.arange(len(all_classes_new)))

ax.set_xticklabels(all_classes_new)
ax.set_yticklabels(all_classes_new)


plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

thresh = cmn.max() / 2.
for i in range(len(all_classes_new)):
    for j in range(len(all_classes_new)):
        text = ax.text(j, i, '{0:.2f}'.format(cmn[i, j]),
                       ha="center", va="center",
                       color="w" if (cmn[i, j] > thresh)or(cmn[i, j]==0.) else "k")

ax.set_ylabel('Labelled')
ax.set_xlabel('Predicted')
ax.set_ylim(len(all_classes_new)-0.5, -0.5)
fig.tight_layout()
```

```python
class_ii = cm.diagonal()
user = np.zeros(class_ii.shape)
user[class_ii!=0] = class_ii[class_ii!=0]/cm[:, class_ii!=0].sum(axis=0)
prod = np.zeros(class_ii.shape)
prod[class_ii!=0] = class_ii[class_ii!=0]/cm[class_ii!=0, :].sum(axis=1)

print(pd.DataFrame(
    data={'class':all_classes_new, 
          'User': user,
          'Prod': prod
         }))
print('Overall = {}'.format(class_ii.sum()/cm.sum()))
```
